### Note on type conversion

Apparently converting from `float` to `int` truncates the float and does not round the value (i.e. all values below `1.0` are truncated to `0` even `0.9` will become `0`). Therefore converting all `NaN` values to 0 first is not a good idea. 

* Perhaps round the numbers first?
    - this will help all values greater than 0.5, but we would lose age data for any babies less than half a year old (only one in the is data set but it still could be a problem)
* Change `NaN` values to -1 or -99?
    - this could work and would distinguish the old nan values, but it could be confusing and may cause problems if try to do arithmetic (i.e. `sum()`) on the column values
* use the `.astype` argument `errors='ignore'` 
    - this will keep `NaN` values as `NaN` but the series/df won't be converted to `int`. It will still be of type `float64`!

```python
#Work on your solution here and after. 
#make sure to follow the given structure below

#1) Write your function here
def data_preprocessing (dataframe=titanic):
    """Write a function named "data_preprocessing" that accepts the dataset in dataframe as only argument and does the following tasks and returns the updated dataframe:
    0) Drop "PassengerId" and "Ticket" columns from the dataframe.
    1) Check if there are any rows that have missing entries for all columns.
    2) Remove such rows from the given dataset.
    3) Convert the data type of "Fare" and "Age" to int.
    4) Rename "SibSp" as "Siblings/Spouses Aboard", and "Parch" as "Parents/Children Aboard".
    5) Under column "Name": replace 'Mlle' with 'Miss', replace 'Ms' with 'Miss' and replace 'Mme'with 'Mrs'."""
    
    df = dataframe.copy() # make a copy so original df not altered
    
    # 0) Drop "PassengerId" and "Ticket" columns from the dataframe.
    df.drop(labels=["PassengerId", "Ticket"], axis=1, inplace=True)
    
    # 1) Check if there are any rows that have missing entries for all columns.
    # 2) Remove such rows from the given dataset.
    nan_rows = []
    for i in range(len(df['Name'])):
        if df.iloc[i].isna().all():
            nan_rows.append(i)
    print("Full NaN rows removed:", nan_rows)
    df.drop(labels=nan_rows, axis=0, inplace=True)
    print(f"The DataFrame now contains {len(df)} rows.")
    
    #3) Convert the data type of "Fare" and "Age" to int.
    df.fillna({'Age':-99, 'Fare':-99}, inplace=True)
    df['Age'] = df['Age'].astype(int)
    df['Fare'] = df['Fare'].round(0)
    df['Fare'] = df['Fare'].astype(int)
    print("Data type of 'Age':", df['Age'].dtypes)
    print("Data type of 'Fare':", df['Fare'].dtypes)
    
    # 4) Rename "SibSp" as "Siblings/Spouses Aboard", and "Parch" as "Parents/Children Aboard".
    df.rename(columns={"SibSp":"Siblings/Spouses Aboard", "Parch":"Parents/Children Aboard"}, inplace=True)
    
    # 5) Under column "Name": replace 'Mlle' with 'Miss', replace 'Ms' with 'Miss' and replace 'Mme'with 'Mrs'.
    df['Name'] = df[['Name']].applymap(lambda x: x.replace('Mlle','Miss')) 
    df['Name'] = df[['Name']].applymap(lambda x: x.replace('Ms','Miss'))
    df['Name'] = df[['Name']].applymap(lambda x: x.replace('Mme','Mrs'))
    
    return df
```

## Question 2:
We are interested in adding new columns which are so-called features.<br/>
Create a new function called "feature_extraction" which does the following updates to the existing dataframe.<br/>
In short, we will fully change the columns "Name", "Embarked", "Sex", "Cabin" and "Age" into numeric.

1. Name column should be of int type.  For those titles in name column, this column should indicate 1 for  "Mr."; 2 for "Miss."; 3 for "Mrs."; 4 for "Master."; 5 for "Rare" and 0 for any others.  
2. Convert "Embarked" column into numeric. Replace "S" with 0; replace "C" with 1 and replace "Q" with 2. For the missing ones (if any), just fill with the most common one.  
3. Convert "Sex" column into numeric.Replace "male" with 0 and replace "female" with 1. For the missing ones (if any), just fill with the most common one.  
4. The column named "Cabin" includes strings with letters and numbers, for example "C85" and the letter refers to the cabin group. This column should indicate integer value of 1 for cabin name starts with letter "A". The rest of the numbers should appear according this dictionary: {"A": 1, "B": 2, "C": 3, "D": 4, "E": 5, "F": 6, "G": 7, "U": 8}. Any missing entries should be replaced with 0.  
5. Fill missing values from "Age" column according to the following ad-hoc imputation technique: A random integer withdrawn from the set (mean - standard deviation, mean + standard deviation).  
6. Any missing values from "Fare" column should be replaced with 0.  
7. Update "Fare" column according to the following:  
    * if 0 <= 'Fare' < 10, then 'Fare' = 0<br/>
    * if 10 <= 'Fare' < 20, then 'Fare' = 1<br/>
    * if 20 <= 'Fare' < 30, then 'Fare' = 2<br/>
    * if 30 <= 'Fare' < 100, then 'Fare' = 3<br/>
    * if 100 <= 'Fare' < 200, then 'Fare' = 4<br/>
    * if 200 <= 'Fare' then 'Fare' = 5<br/>
8. Update "Age" column according to the following:
   * if 'Age' <= 10, then 'Age' = 0
   * if 'Age' > 10 & 'Age' <= 15 then 'Age' = 1
   * if 'Age' > 15 & 'Age' <= 20 then 'Age' = 2
   * if 'Age' > 20 & 'Age' <= 25 then 'Age' = 3
   * if 'Age' > 25 & 'Age' <= 35 then 'Age' = 4 . . . *(is this a purposeful 10 year gap?)*
   * if 'Age' > 35 & 'Age' <= 40 then 'Age' = 5
   * if 'Age' > 40 & 'Age' <= 60 then 'Age' = 6
   * if 'Age' > 60 then 'Age' = 7

#### Note:
For part 1 of this question instead of "Rare", I added groups for:

"Rev." = 5  
"Dr." = 6  
"other" = 0 (and includes Military and Nobility titles)

```python
#Work on your solution here and after. 
#1) Write your function here
def feature_extraction (df):
    #body of the function
    
    # 1. Convert titles in name column to:
    titles = {"Mr.":1, "Miss.":2, "Mrs.":3, "Master.":4, "Rev.":5, "Dr.":6, "other":0}
    
    df.insert(2, "Title", value="other")  # create new col to collect title info
    df.at[df['Name'].str.contains("Mr\."),'Title'] = "Mr."
    df.at[df['Name'].str.contains("Miss\."),'Title'] = "Miss."
    df.at[df['Name'].str.contains("Mrs\."),'Title'] = "Mrs."
    df.at[df['Name'].str.contains("Master\."),'Title'] = "Master."
    df.at[df['Name'].str.contains("Rev\."),'Title'] = "Rev."
    df.at[df['Name'].str.contains("Dr\."),'Title'] = "Dr."
    
    df['Name'] = df.Title.map(titles)
    df.drop('Title', axis=1, inplace=True)  # remove extra col
    df['Name'] = df['Name'].astype(int)
    
    
    # 2. Convert "Embarked" column into numeric. Replace "S" with 0; replace "C" with 1 and replace "Q" with 2. 
    # For the missing ones (if any), just fill with the most common one.
    df['Embarked'] = df.Embarked.map({"S":0, "C":1, "Q":2})
    df['Embarked'].fillna(df['Embarked'].value_counts().idxmax(), inplace=True)
    df['Embarked'] = df['Embarked'].astype(int)
    
    # 3. Convert "Sex" column into numeric.Replace "male" with 0 and replace "female" with 1. 
    # For the missing ones (if any), just fill with the most common one.
    df['Sex'] = df.Sex.map({'male':0, 'female':1})
    df['Sex'].fillna(df['Sex'].value_counts().idxmax(), inplace=True)
    df['Sex'] = df['Sex'].astype(int)
    
    # 4. The column named "Cabin" includes strings with letters and numbers. This column should indicate integer value 
    # according this dictionary: {"A": 1, "B": 2, "C": 3, "D": 4, "E": 5, "F": 6, "G": 7, "U": 8}. 
    # Any missing entries should be replaced with 0.
    df.insert(9, column='Cabin1st', value=df['Cabin'])   # insert a new col to transform
    df['Cabin1st'].fillna("00",inplace=True)             # replace NaN with string so series becomes subccriptable
    df['Cabin1st'] = df.Cabin1st.map(lambda x: x[0],)    # keep only 1st character of string
    
    cabin_dict = {"A": 1, "B": 2, "C": 3, "D": 4, "E": 5, "F": 6, "G": 7, "U": 8}
    df['Cabin1st'] = df.Cabin1st.map(cabin_dict)
    df['Cabin1st'].fillna(0,inplace=True)           # fill NaN values reintroduced by map function
    df['Cabin'] = df['Cabin1st'].astype(int)
    df.drop('Cabin1st', axis=1, inplace=True)
    
    
    # 5. Fill missing values from "Age" column according to the following ad-hoc imputation technique: 
    # A random integer withdrawn from the set (mean - standard deviation, mean + standard deviation).
    age_mean = df.loc[~(df['Age'] == -99), 'Age'].mean()
    age_std = df.loc[~(df['Age'] == -99), 'Age'].std()
    age_list = list(range((int(age_mean - age_std)+1),(int(age_mean + age_std)+1)))

    import random
    df.loc[df['Age']==-99, 'Age'] = random.choices(age_list, k=sum(df['Age']==-99))
    
    
    # 6. Any missing values from "Fare" column should be replaced with 0.
    df.loc[df['Fare']==-99, 'Fare'] = 0
    
    # 7. Update "Fare" column according to:
    for f in df['Fare'].index:
        if df.loc[f, 'Fare'] < 10:
            df.loc[f, 'Fare'] = 0
        elif df.loc[f, 'Fare'] < 20:
            df.loc[f, 'Fare'] = 1
        elif df.loc[f, 'Fare'] < 30:
            df.loc[f, 'Fare'] = 2
        elif df.loc[f, 'Fare'] < 100:
            df.loc[f, 'Fare'] = 3
        elif df.loc[f, 'Fare'] < 200:
            df.loc[f, 'Fare'] = 4
        else:
            df.loc[f, 'Fare'] = 5
    
    # 8. Update "Age" column according to:
    for f in df['Age'].index:
        if df.loc[f, 'Age'] <= 10:
            df.loc[f, 'Age'] = 0
        elif df.loc[f, 'Age'] <= 15:
            df.loc[f, 'Age'] = 1
        elif df.loc[f, 'Age'] <= 20:
            df.loc[f, 'Age'] = 2
        elif df.loc[f, 'Age'] <= 25:
            df.loc[f, 'Age'] = 3
        elif df.loc[f, 'Age'] <= 35:
            df.loc[f, 'Age'] = 4
        elif df.loc[f, 'Age'] <= 40:
            df.loc[f, 'Age'] = 5
        elif df.loc[f, 'Age'] <= 60:
            df.loc[f, 'Age'] = 6
        else:
            df.loc[f, 'Age'] = 7
    
    return df

#2) Call your function here
new_titanic = feature_extraction(new_titanic)
new_titanic.head()
```

### Question 2
Only looking at the three most populous counties for each state, what are the three most populous states (in order of highest population to lowest population)? Use `CENSUS2010POP`.

*This function should return a list of string values.*

```python
def statepop_top3(dataframe=census_df):
    """This functuon returns the top 3 states with the largest populations according to the total state populations"""
    states = census_df[census_df['SUMLEV'] == 40] # create new df with only states data
    return list(states.sort_values('CENSUS2010POP',ascending = False).head(3)['STNAME']) 
```

```python
statepop_top3(census_df)
```

```python
def statepop_top4_by3topcty(dataframe=census_df):
    """This functuon returns the top 4 states with the largest populations according to 
    each state's top 3 most populous counties"""
    counties = census_df[census_df['SUMLEV'] == 50]
    top3 = counties.groupby('STNAME')['CENSUS2010POP'].nlargest(3).index
    top3_index = [i[1] for i in top3]
    statepop_bytop3 = counties.loc[top3_index].groupby('STNAME').sum()
    return list(statepop_bytop3.sort_values('CENSUS2010POP',ascending = False).head(4).index)
```

### Question 3
Which county has had the largest absolute change in population within the period 2010-2015? (Hint: population values are stored in columns POPESTIMATE2010 through POPESTIMATE2015, you need to consider all six columns.)

e.g. If County Population in the 5 year period is 100, 120, 80, 105, 100, 130, then its largest change in the period would be |130-80| = 50.

*This function should return a single string value.*

```python
def popestimate_max(df=census_df):
    """ Determines the county that had the largest absolute change in population within the period 2010-2015"""
    counties = census_df[census_df['SUMLEV'] == 50]
    popestimate = counties.loc[:, "CTYNAME":"POPESTIMATE2015"]
    popestimate.set_index('CTYNAME', inplace=True)
    popestimate.drop(['CENSUS2010POP','ESTIMATESBASE2010'], axis=1, inplace=True)
    return popestimate.apply(lambda x: abs(x.max()-x.min()), axis=1).idxmax()    
```

```python
popestimate_max(census_df)
```

### Question 4
In this datafile, the United States is broken up into four regions using the "REGION" column. 

Create a query that finds the counties that belong to regions 1 or 2, whose name starts with 'Washington', and whose POPESTIMATE2015 was greater than their POPESTIMATE 2014.

*This function should return a 5x2 DataFrame with the columns = ['STNAME', 'CTYNAME'] and the same index ID as the census_df (sorted ascending by index).*

```python
def washington_counties(dataframe=census_df):
    """Creates a 5x2 dataframe with all county names that: 
    1. start with 'Washington'
    2. are in regions 1 or 2
    3. whose POPESTIMATE2015 was greater than their POPESTIMATE2014."""
    counties = census_df[census_df['SUMLEV'] == 50]
    region12 = counties[(counties['REGION'] == 1) | (counties['REGION'] == 2)]
    region12_pop = region12[region12['POPESTIMATE2015'] > region12['POPESTIMATE2014']].filter(['STNAME','CTYNAME'])
    return region12_pop[region12_pop['CTYNAME'].str.contains("^Washington")]
```
