## Question 3:
Modify `data_preprocessing` function so that it receives an object as args. Perform the same data manipulation operations that we did in this function in Assignment-2 for "data". This function should attach the data manipulated results back to my_data_object. 

```python
#Work on your solution here and after. 
#1) Write your function here
def data_preprocessing (obj):
    #body of the function
    df = obj.data #get the complete dataframe from the object.
    #0) Drop "PassengerId" and "Ticket" columns from the dataframe.
    df = df.drop(labels = ['PassengerId', 'Ticket'], axis = 1) #drop the columns

    #1) Check if there are any rows that have missing entries for all columns.   
    #we will use any() function to check a list is full of False
    #any function returns False only if all elements are false in an iterable object
    #we will collect each row of the dataframe within a True/False list and check with any()
    #If all false, then any() will return false
    bool_df = ~df.isna() #True means there is a value. False means the element is na.
    #check each row and mark it as false if there is any rows with all columns false
    #we will use Pandas.apply() function to each row
    ps = bool_df.apply(any, axis = 1) #this is a pandas series of type bool. False if all rows are false.

    #2) Remove such rows from the given dataset.
    new_df = df[ps] #grab the ones with True rows only

    #3) Convert the data type of "Fare" and "Age" to int.
    #There may be NaNs in grids of the dataframe.
    ps_na1 = new_df.Fare.isna()
    ps_na2 = new_df.Age.isna()
    #Let's convert those values to -1 as there is no integer NaN defined in Python.
    if any(ps_na1): #if there is any na in Fare column
        new_df.loc[ps_na1,['Fare']] = -1
    if any(ps_na2): #if there is any na in Age column
        new_df.loc[ps_na2,['Age']] = -1
    #round values under "Fare" column
    new_df.loc[:,['Fare']] = new_df.loc[:,['Fare']].round()
    new_df.loc[:,['Fare', 'Age']] = new_df.loc[:,['Fare', 'Age']].astype(int)

    #4) Rename "SibSp" as "Siblings/Spouses Aboard", and "Parch" as "Parents/Children Aboard".
    new_df = new_df.rename(columns={"SibSp": "Siblings/Spouses Aboard", "Parch": "Parents/Children Aboard"})

    #5) Under column "Name": replace 'Mlle' with 'Miss', replace 'Ms' with 'Miss' and replace 'Mme'with 'Mrs'.
    new_df['Name'] = new_df['Name'].str.replace('Mlle','Miss')
    new_df['Name'] = new_df['Name'].str.replace('Ms','Miss')
    new_df['Name'] = new_df['Name'].str.replace('Mme','Mrs')

    obj.data = new_df #Attach the updated dataframe
    return obj #return your object



#2) Call your function here
my_data_object = data_preprocessing(my_data_object)
my_data_object.data.head() #report the head of train dataframe
```

## Question 4:
Modify `feature_extraction` function so that it receives an object as args. Perform the same data manipulation operations that we did in this function function in Assignment-2 for "data". This function should attach the data manipulated results back to my_data_object. 

```python
#Work on your solution here and after. 
#1) Write your function here
def feature_extraction (obj): 
    #body of the function
    df = obj.data #get the complete dataframe from the object.

    #1) Name column should be of int type. 
    #For those titles in name column, 
    #this column should indicate 1 for "Mr."; 2 for "Miss."; 3 for "Mrs."; 4 for "Master."; 5 for "Rare" 
    #and 0 for any others.

    my_titles_dict = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Rare": 5} #we'll use this dict in our .map() function.
    #let's create a new column called "Title" and extract all title from Name column
    #We will use pd.Series.str.extrace() function.
    df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand = False)
    #If you don't see the above given function, you should study this function from the docs.
    #Here is a simple example for setting a pattern to look for within a string.
    #This could be a simple example to see it.
    #pd.Series(['abc a3. ', 'def b4.', 'ghi c5. ']).str.extract(' ([A-Za-z0-9]+)\.') 

    # convert titles into numbers
    df['Title'] = df['Title'].map(my_titles_dict)
    # Convert NaNs to 0 (if any)
    df['Title'] = df['Title'].fillna(0)
    # Convert data type to integer
    df['Title'] = df['Title'].astype(int)
    df[['Name']] = df[['Title']].copy()
    df.drop(['Title'], axis = 1, inplace = True)


    #2) Convert "Embarked" column into numeric. 
    #Replace "S" with 0; replace "C" with 1 and replace "Q" with 2. 
    #For the missing ones (if any), just fill with the most common one.

    #first, let's fill the missing ones with the most common one
    #the following code reveals the most common one is 'S'
    most_one = df['Embarked'].value_counts().idxmax() 
    df['Embarked'] = df['Embarked'].fillna(most_one)

    #now let's replace the characters with the required digits
    my_dict = {"S": 0, "C": 1, "Q": 2} #we will use this dictionary in map function
    df['Embarked'] = df['Embarked'].map(my_dict)


    #3) Convert "Sex" column into numeric.
    #Replace "male" with 0 and replace "female" with 1. 
    #For the missing ones (if any), just fill with the most common one.

    #first, let's fill the missing ones with the most common one
    #the following code reveals the most common one is 'S'
    #As you may guess, the most common one is male passenger in Titanic.
    most_one = df['Sex'].value_counts().idxmax() 
    df['Sex'] = df['Sex'].fillna(most_one)
    my_dict = {"male": 0, "female": 1}
    df['Sex'] = df['Sex'].map(my_dict)


    #4) The column named "Cabin" includes strings with letters and numbers, 
    #for example "C85" and the letter refers to the cabin group. 
    #This column should indicate integer value of 1 for cabin name starts with letter "A". 
    #The rest of the numbers should appear according this dictionary: 
    #{"A": 1, "B": 2, "C": 3, "D": 4, "E": 5, "F": 6, "G": 7, "U": 8}. 
    #Any missing entries should be replaced with 0.

    my_dict = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5, 'F': 6, 'G': 7, 'U': 8}
    #the following code gives starting letters of each entry under Cabin column
    df['Cabin'] = df['Cabin'].str.extract('([A-Za-z0-9])')

    df['Cabin'] = df['Cabin'].map(my_dict)
    df['Cabin'] = df['Cabin'].fillna(0) #fill NaNs with 0s.
    df['Cabin'] = df['Cabin'].astype(int) #convert data type to integer

    #5) Fill missing values from "Age" column according to the following ad-hoc imputation technique: 
    #A random integer withdrawn from the set (mean - standard deviation, mean + standard deviation).


    #Remember we replaced NaNs with -1s in Question-1. 
    #Let's load back NaNs so that mean and standard deviation won't get affected
    ps_age = df['Age'].replace(-1, float('NaN'))
    #which ones are NaNs
    na_age = ps_age.isna()
    #mean and std of all ages
    mean_age = ps_age.mean()
    std_age = ps_age.std()
    #how many NaNs
    size_na_age = sum(na_age)

    #Let's use Numpy package to randomly withdraw integer values from the given range 
    #(mean - standard deviation, mean + standard deviation) as many as size
    import numpy as np
    #randint() gives an array which can be inserted in Pandas series
    ps_rand_age_array = np.random.randint(max(0, mean_age - std_age), mean_age + std_age, size = size_na_age)
    #Only replace for those NaNs.
    ps_age[na_age] = ps_rand_age_array
    #Make sure it is int type again.
    ps_age = ps_age.astype(int)
    df['Age'] = ps_age

    #6) Any missing values from "Fare" column should be replaced with 0.
    #Remember we replaced NaNs with -1s in Question-1. 
    #Let's load back NaNs so that mean and standard deviation won't get affected
    ps_fare = df['Fare'].replace(-1, float('NaN'))
    #which ones are NaNs
    na_fare = ps_fare.isna()
    #Only replace for those NaNs.
    ps_fare[na_fare] = 0
    #Make sure it is int type again.
    ps_fare = ps_fare.astype(int)
    df['Fare'] = ps_fare


    # 7) Update "Fare" column according to the following:
    # if 0 <= 'Fare' < 10, then 'Fare' = 0
    # if 10 <= 'Fare' < 20, then 'Fare' = 1
    # if 20 <= 'Fare' < 30, then 'Fare' = 2
    # if 30 <= 'Fare' < 100, then 'Fare' = 3
    # if 100 <= 'Fare' < 200, then 'Fare' = 4
    # if 200 <= 'Fare' then 'Fare' = 5
    df.loc[ df['Fare'] < 10, 'Fare'] = 0
    df.loc[(df['Fare'] >= 10) & (df['Fare'] < 20), 'Fare'] = 1
    df.loc[(df['Fare'] >= 20) & (df['Fare'] < 30), 'Fare'] = 2
    df.loc[(df['Fare'] >= 30) & (df['Fare'] < 100), 'Fare'] = 3
    df.loc[(df['Fare'] >= 100) & (df['Fare'] < 200), 'Fare'] = 4
    df.loc[(df['Fare'] >= 200), 'Fare'] = 5


    # 8) Update "Age" column according to the following:
    # if 'Age' <= 10, then 'Age' = 0
    # if 'Age' > 10 & 'Age' <= 15 then 'Age' = 1
    # if 'Age' > 15 & 'Age' <= 20 then 'Age' = 2
    # if 'Age' > 20 & 'Age' <= 25 then 'Age' = 3
    # if 'Age' > 25 & 'Age' <= 35 then 'Age' = 4
    # if 'Age' > 35 & 'Age' <= 40 then 'Age' = 5
    # if 'Age' > 40 & 'Age' <= 60 then 'Age' = 6
    # if 'Age' > 60 then 'Age' = 6
    df.loc[ df['Age'] <= 10, 'Age'] = 0
    df.loc[(df['Age'] > 10) & (df['Age'] <= 15), 'Age'] = 1
    df.loc[(df['Age'] > 15) & (df['Age'] <= 20), 'Age'] = 2
    df.loc[(df['Age'] > 20) & (df['Age'] <= 25), 'Age'] = 3
    df.loc[(df['Age'] > 25) & (df['Age'] <= 35), 'Age'] = 4
    df.loc[(df['Age'] > 35) & (df['Age'] <= 40), 'Age'] = 5
    df.loc[(df['Age'] > 40) & (df['Age'] <= 60), 'Age'] = 6
    df.loc[ df['Age'] > 60, 'Age'] = 6

    obj.data = df
    return obj


#2) Call your function here
#You must call the data_preprocessing() function first
my_data_object = feature_extraction(my_data_object) 
my_data_object.data.head() #report the head of train dataframe
```
