### Question 7:
Rename column named "Height" to "Height (feet-inches)". Add a new column named "Height (inches)" to the existing dataframe. The column should include values in total inches. Say, for example, if a player's height is 6-0, the new column should show it as 72. Add another column named "Tall" to the existing dataframe. This column should indicate a bool type value of True if the associated player height is greater than or equal to 80 inches. Otherwise, its value should be False. Use for loop and if statement in your solution.<br/>

### Solution 7:

```python
#Work on your solution here and after. 
nba_new = nba.copy() #clone an exact copy
myListFeetInches = []
myListTallShort = []

for element in nba_new['Height']:
    k = str(element) #create a string out of object
    if k!='nan':
        feet = k[0]
        inches = k[2:]
        feetInches = 12*int(feet)+ int(inches)
        if feetInches >= 80:
            playerHeight = True #it is a tall player
        else:
            playerHeight = False
        myListTallShort.append(playerHeight)
        myListFeetInches.append(feetInches)
    else: #if the entry is 'nan'
        myListFeetInches.append(float('nan'))
        myListTallShort.append(bool('nan')) #make sure it is boolean defined


nba_new[['Height (inches)']] =  pd.DataFrame(myListFeetInches) #create the new column with the associated list
nba_new[['Tall']] = pd.DataFrame(myListTallShort)
#nba_new['Height (inches)'].dtype #check the datatype of the column
print(nba_new.Tall.dtype) #print new column type
nba_new.head() #updated dataframe first 5 rows

```

### Question 8:

### Question 9:
Add a new column named "Young" to the existing dataframe. This column should indicate a bool type value of True if the associated player age is less than 30. Otherwise, its value should be False. Use for loop and if statement in your solution.<br/>

### Solution 9:

```python
#Work on your solution here and after. 
nba_new = nba.copy() #clone an exact copy
myListYoungOld = []

for age in nba_new['Age']:
    if age!='nan':
        if age < 30:
            young_old = True #it is a young player
        else:
            young_old = False
        myListYoungOld.append(young_old)
    else: #if the entry is 'nan'
        myListYoungOld.append(bool('nan')) #make sure it is boolean defined

nba_new[['Young']] = pd.DataFrame(myListYoungOld)
print(nba_new.Young.dtype) #print new column type
nba_new.head(40) #updated dataframe first 5 rows
```
