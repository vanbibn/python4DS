```python
#Fill in this class
import pandas as pd
class My_data_class:
    def __init__(self):
        self.data = pd.DataFrame([1,2]) #default data type
        self.train = pd.DataFrame([1]) #default data type
        self.test = pd.DataFrame([2]) #default data type
            
    def create_test_train(self): #create a method to find test and train data
        # create a random sample of 20% of the data to test
        data = self.data
        test_data = data.sample(frac=0.2, random_state=48, axis=0)
        self.test = test_data
        
        # create a dataframe with the remaining 80%
        data_index = data.index.tolist()
        test_index = test_data.index.tolist()
        train_index = list(set(data_index)-set(test_index))
        train_data = data.loc[train_index]
        self.train = train_data
    
        
    @property #how we define a getter. Always define a getter before a setter function.
    def data(self): #a getter function, which is a method in this class. 
        return self.__data
    
    @data.setter #how we define a setter
    def data(self, my_data): #a setter function, which is a method in this class.
        if isinstance(my_data, pd.DataFrame):
            self.__data = my_data #no "return" is necessary
        else:
            raise ValueError(f"{my_data} is not a Pandas DataFrame.")
    
    # getter and setter for train
    @property
    def train(self):
        return self.__train
    
    @train.setter
    def train(self, my_data):
        if isinstance(my_data, pd.DataFrame):
            self.__train = my_data
        else:
            raise ValueError(f"{my_data} is not a Pandas DataFrame.")
        
    # getter and setter for test
    @property
    def test(self):
        return self.__test
    
    @test.setter
    def test(self, my_data):
        if isinstance(my_data, pd.DataFrame):
            self.__test = my_data
        else:
            raise ValueError(f"{my_data} is not a Pandas DataFrame.")
        
    def __str__(self):
        return "This is a data object"
```

## Question 3:
Modify `data_preprocessing` function so that it receives an object as args. Perform the same data manipulation operations that we did in this function in Assignment-2 for "data". This function should attach the data manipulated results back to my_data_object. 

```python
#Work on your solution here and after. 
#1) Write your function here
def data_preprocessing (obj):
    """Write a function named "data_preprocessing" that accepts the dataset in dataframe as only argument and does the following tasks and returns the updated dataframe:
    0) Drop "PassengerId" and "Ticket" columns from the dataframe.
    1) Check if there are any rows that have missing entries for all columns.
    2) Remove such rows from the given dataset.
    3) Convert the data type of "Fare" and "Age" to int.
    4) Rename "SibSp" as "Siblings/Spouses Aboard", and "Parch" as "Parents/Children Aboard".
    5) Under column "Name": replace 'Mlle' with 'Miss', replace 'Ms' with 'Miss' and replace 'Mme'with 'Mrs'."""
    
    df = obj.data #.copy() # make a copy so original df not altered
    
    # 0) Drop "PassengerId" and "Ticket" columns from the dataframe.
    df.drop(labels=["PassengerId", "Ticket"], axis=1, inplace=True)
    
    # 1) Check if there are any rows that have missing entries for all columns.
    # 2) Remove such rows from the given dataset.
    nan_rows = []
    for i in range(len(df['Name'])):
        if df.iloc[i].isna().all():
            nan_rows.append(i)
    print("Full NaN rows removed:", nan_rows)
    df.drop(labels=nan_rows, axis=0, inplace=True)
    print(f"The DataFrame now contains {len(df)} rows.")
    
    #3) Convert the data type of "Fare" and "Age" to int.
    df.fillna({'Age':-99, 'Fare':-99}, inplace=True)
    df['Age'] = df['Age'].astype(int)
    df['Fare'] = df['Fare'].round(0)
    df['Fare'] = df['Fare'].astype(int)
    print("Data type of 'Age':", df['Age'].dtypes)
    print("Data type of 'Fare':", df['Fare'].dtypes)
    
    # 4) Rename "SibSp" as "Siblings/Spouses Aboard", and "Parch" as "Parents/Children Aboard".
    df.rename(columns={"SibSp":"Siblings/Spouses Aboard", "Parch":"Parents/Children Aboard"}, inplace=True)
    
    # 5) Under column "Name": replace 'Mlle' with 'Miss', replace 'Ms' with 'Miss' and replace 'Mme'with 'Mrs'.
    df['Name'] = df[['Name']].applymap(lambda x: x.replace('Mlle','Miss')) 
    df['Name'] = df[['Name']].applymap(lambda x: x.replace('Ms','Miss'))
    df['Name'] = df[['Name']].applymap(lambda x: x.replace('Mme','Mrs'))
    
    # reassign processed df to data attribute of the object
    obj.data = df
    
    return obj



#2) Call your function here
my_data_object = data_preprocessing(my_data_object)
my_data_object.data.head() #report the head of train dataframe
```

```python
#Work on your solution here and after. 
#1) Write your function here
def feature_extraction (obj): 
    #body of the function
    return obj #return your object

def feature_extraction (obj):
    #body of the function
    df = obj.data#.copy() # make a copy so original df not altered
    
    # 1. Convert titles in name column to:
    titles = {"Mr.":1, "Miss.":2, "Mrs.":3, "Master.":4, "Rev.":5, "Dr.":6, "other":0}
    
    df.insert(2, "Title", value="other")  # create new col to collect title info
    df.at[df['Name'].str.contains("Mr\."),'Title'] = "Mr."
    df.at[df['Name'].str.contains("Miss\."),'Title'] = "Miss."
    df.at[df['Name'].str.contains("Mrs\."),'Title'] = "Mrs."
    df.at[df['Name'].str.contains("Master\."),'Title'] = "Master."
    df.at[df['Name'].str.contains("Rev\."),'Title'] = "Rev."
    df.at[df['Name'].str.contains("Dr\."),'Title'] = "Dr."
    
    df['Name'] = df.Title.map(titles)
    df.drop('Title', axis=1, inplace=True)  # remove extra col
    df['Name'] = df['Name'].astype(int)
    
    
    # 2. Convert "Embarked" column into numeric. Replace "S" with 0; replace "C" with 1 and replace "Q" with 2. 
    # For the missing ones (if any), just fill with the most common one.
    df['Embarked'] = df.Embarked.map({"S":0, "C":1, "Q":2})
    df['Embarked'].fillna(df['Embarked'].value_counts().idxmax(), inplace=True)
    df['Embarked'] = df['Embarked'].astype(int)
    
    # 3. Convert "Sex" column into numeric.Replace "male" with 0 and replace "female" with 1. 
    # For the missing ones (if any), just fill with the most common one.
    df['Sex'] = df.Sex.map({'male':0, 'female':1})
    df['Sex'].fillna(df['Sex'].value_counts().idxmax(), inplace=True)
    df['Sex'] = df['Sex'].astype(int)
    
    # 4. The column named "Cabin" includes strings with letters and numbers. This column should indicate integer value 
    # according this dictionary: {"A": 1, "B": 2, "C": 3, "D": 4, "E": 5, "F": 6, "G": 7, "U": 8}. 
    # Any missing entries should be replaced with 0.
    df.insert(9, column='Cabin1st', value=df['Cabin'])   # insert a new col to transform
    df['Cabin1st'].fillna("00",inplace=True)             # replace NaN with string so series becomes subccriptable
    df['Cabin1st'] = df.Cabin1st.map(lambda x: x[0],)    # keep only 1st character of string
    
    cabin_dict = {"A": 1, "B": 2, "C": 3, "D": 4, "E": 5, "F": 6, "G": 7, "U": 8}
    df['Cabin1st'] = df.Cabin1st.map(cabin_dict)
    df['Cabin1st'].fillna(0,inplace=True)           # fill NaN values reintroduced by map function
    df['Cabin'] = df['Cabin1st'].astype(int)
    df.drop('Cabin1st', axis=1, inplace=True)
    
    
    # 5. Fill missing values from "Age" column according to the following ad-hoc imputation technique: 
    # A random integer withdrawn from the set (mean - standard deviation, mean + standard deviation).
    age_mean = df.loc[~(df['Age'] == -99), 'Age'].mean()
    age_std = df.loc[~(df['Age'] == -99), 'Age'].std()
    age_list = list(range((int(age_mean - age_std)+1),(int(age_mean + age_std)+1)))

    import random
    df.loc[df['Age']==-99, 'Age'] = random.choices(age_list, k=sum(df['Age']==-99))
    
    
    # 6. Any missing values from "Fare" column should be replaced with 0.
    df.loc[df['Fare']==-99, 'Fare'] = 0
    
    # 7. Update "Fare" column according to:
    for f in df['Fare'].index:
        if df.loc[f, 'Fare'] < 10:
            df.loc[f, 'Fare'] = 0
        elif df.loc[f, 'Fare'] < 20:
            df.loc[f, 'Fare'] = 1
        elif df.loc[f, 'Fare'] < 30:
            df.loc[f, 'Fare'] = 2
        elif df.loc[f, 'Fare'] < 100:
            df.loc[f, 'Fare'] = 3
        elif df.loc[f, 'Fare'] < 200:
            df.loc[f, 'Fare'] = 4
        else:
            df.loc[f, 'Fare'] = 5
    
    # 8. Update "Age" column according to:
    for f in df['Age'].index:
        if df.loc[f, 'Age'] <= 10:
            df.loc[f, 'Age'] = 0
        elif df.loc[f, 'Age'] <= 15:
            df.loc[f, 'Age'] = 1
        elif df.loc[f, 'Age'] <= 20:
            df.loc[f, 'Age'] = 2
        elif df.loc[f, 'Age'] <= 25:
            df.loc[f, 'Age'] = 3
        elif df.loc[f, 'Age'] <= 35:
            df.loc[f, 'Age'] = 4
        elif df.loc[f, 'Age'] <= 40:
            df.loc[f, 'Age'] = 5
        elif df.loc[f, 'Age'] <= 60:
            df.loc[f, 'Age'] = 6
        else:
            df.loc[f, 'Age'] = 7
    
    # reassign processed df to data attribute of the object
    obj.data = df
    
    return obj
    

#2) Call your function here
my_data_object = feature_extraction(my_data_object)
my_data_object.data.head() #report the head of train dataframe
```
