### Question 1:
Write a function named "data_preprocessing" that accepts the dataset in dataframe as only argument and does the following tasks and returns the updated dataframe:<br/>
0) Drop "PassengerId" and "Ticket" columns from the dataframe.<br/>
1) Check if there are any rows that have missing entries for all columns.<br/> 
2) Remove such rows from the given dataset.<br/>
3) Convert the data type of "Fare" and "Age" to int.<br/>
4) Rename "SibSp" as "Siblings/Spouses Aboard", and "Parch" as "Parents/Children Aboard".<br/>
5) Under column "Name": replace 'Mlle' with 'Miss', replace 'Ms' with 'Miss' and replace 'Mme' with 'Mrs'.<br/>

```python
#Work on your solution here and after. 
#make sure to follow the given structure below

#Write your function here
def data_preprocessing (args):
    #body of the function
    #0) Drop "PassengerId" and "Ticket" columns from the dataframe.
    df = args.drop(labels = ['PassengerId', 'Ticket'], axis = 1) #drop the columns
    df.head()

    #1) Check if there are any rows that have missing entries for all columns.   
    #we will use any() function to check a list is full of False
    #any function returns False only if all elements are false in an iterable object
    #we will collect each row of the dataframe within a True/False list and check with any()
    #If all false, then any() will return false
    bool_df = ~df.isna() #True means there is a value. False means the element is na.
    #check each row and mark it as false if there is any rows with all columns false
    #we will use Pandas.apply() function to each row
    ps = bool_df.apply(any, axis = 1) #this is a pandas series of type bool. False if all rows are false.

    #2) Remove such rows from the given dataset.
    new_df = df[ps] #grab the ones with True rows only

    #3) Convert the data type of "Fare" and "Age" to int.
    #There may be NaNs in grids of the dataframe.
    ps_na1 = new_df.Fare.isna()
    ps_na2 = new_df.Age.isna()
    #Let's convert those values to -1 as there is no integer NaN defined in Python.
    if any(ps_na1): #if there is any na in Fare column
        new_df.loc[ps_na1,['Fare']] = -1
    if any(ps_na2): #if there is any na in Age column
        new_df.loc[ps_na2,['Age']] = -1
    #round values under "Fare" column
    new_df.loc[:,['Fare']] = new_df.loc[:,['Fare']].round()
    new_df.loc[:,['Fare', 'Age']] = new_df.loc[:,['Fare', 'Age']].astype(int)

    #4) Rename "SibSp" as "Siblings/Spouses Aboard", and "Parch" as "Parents/Children Aboard".
    new_df = new_df.rename(columns={"SibSp": "Siblings/Spouses Aboard", "Parch": "Parents/Children Aboard"})

    #5) Under column "Name": replace 'Mlle' with 'Miss', replace 'Ms' with 'Miss' and replace 'Mme'with 'Mrs'.
    new_df['Name'] = new_df['Name'].str.replace('Mlle','Miss')
    new_df['Name'] = new_df['Name'].str.replace('Ms','Miss')
    new_df['Name'] = new_df['Name'].str.replace('Mme','Mrs')

    return (new_df)

#Call your function here
new_titanic = data_preprocessing(titanic)
new_titanic.head()
```

## Question 2:
We are interested in adding new columns which are so-called features.<br/>
Create a new function called "feature_extraction" which does the following updates to the existing dataframe.<br/>
In short, we will fully change the columns "Name", "Embarked", "Sex", "Cabin" and "Age" into numeric.<br/>
1) Name column should be of int type.  For those titles in name column, this column should indicate 1 for  "Mr."; 2 for "Miss."; 3 for "Mrs."; 4 for "Master."; 5 for "Rare" and 0 for any others.<br/>
2) Convert "Embarked" column into numeric. Replace "S" with 0; replace "C" with 1 and replace "Q" with 2. For the missing ones (if any), just fill with the most common one.<br/>
3) Convert "Sex" column into numeric.Replace "male" with 0 and replace "female" with 1. For the missing ones (if any), just fill with the most common one.<br/>
4) The column named "Cabin" includes strings with letters and numbers, for example "C85" and the letter refers to the cabin group. This column should indicate integer value of 1 for cabin name starts with letter "A". The rest of the numbers should appear according this dictionary: {"A": 1, "B": 2, "C": 3, "D": 4, "E": 5, "F": 6, "G": 7, "U": 8}. Any missing entries should be replaced with 0.<br/>
5) Fill missing values from "Age" column according to the following ad-hoc imputation technique: A random integer withdrawn from the set (mean - standard deviation, mean + standard deviation).<br/>
6) Any missing values from "Fare" column should be replaced with 0.<br/>
7) Update "Fare" column according to the following:<br/>
  if 0 <= 'Fare' < 10, then 'Fare' = 0<br/>
  if 10 <= 'Fare' < 20, then 'Fare' = 1<br/>
  if 20 <= 'Fare' < 30, then 'Fare' = 2<br/>
  if 30 <= 'Fare' < 100, then 'Fare' = 3<br/>
  if 100 <= 'Fare' < 200, then 'Fare' = 4<br/>
  if 200 <= 'Fare' then 'Fare' = 5<br/>
  
8) Update "Age" column according to the following:<br/>
   if 'Age' <= 10, then 'Age' = 0<br/>
   if 'Age' > 10 & 'Age' <= 15 then 'Age' = 1<br/>
   if 'Age' > 15 & 'Age' <= 20 then 'Age' = 2<br/>
   if 'Age' > 20 & 'Age' <= 25 then 'Age' = 3<br/>
   if 'Age' > 25 & 'Age' <= 35 then 'Age' = 4<br/>
   if 'Age' > 35 & 'Age' <= 40 then 'Age' = 5<br/>
   if 'Age' > 40 & 'Age' <= 60 then 'Age' = 6<br/>
   if 'Age' > 60 then 'Age' = 6


```python
#Work on your solution here and after. 
#Write your function here
def feature_extraction (args):
    #body of the function
    df = args.copy()

    #1) Name column should be of int type. 
    #For those titles in name column, 
    #this column should indicate 1 for "Mr."; 2 for "Miss."; 3 for "Mrs."; 4 for "Master."; 5 for "Rare" 
    #and 0 for any others.

    my_titles_dict = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Rare": 5} #we'll use this dict in our .map() function.
    #let's create a new column called "Title" and extract all title from Name column
    #We will use pd.Series.str.extrace() function.
    df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand = False)
    #If you don't see the above given function, you should study this function from the docs.
    #Here is a simple example for setting a pattern to look for within a string.
    #This could be a simple example to see it.
    #pd.Series(['abc a3. ', 'def b4.', 'ghi c5. ']).str.extract(' ([A-Za-z0-9]+)\.') 

    # convert titles into numbers
    df['Title'] = df['Title'].map(my_titles_dict)
    # Convert NaNs to 0 (if any)
    df['Title'] = df['Title'].fillna(0)
    # Convert data type to integer
    df['Title'] = df['Title'].astype(int)
    df[['Name']] = df[['Title']].copy()
    df.drop(['Title'], axis = 1, inplace = True)


    #2) Convert "Embarked" column into numeric. 
    #Replace "S" with 0; replace "C" with 1 and replace "Q" with 2. 
    #For the missing ones (if any), just fill with the most common one.

    #first, let's fill the missing ones with the most common one
    #the following code reveals the most common one is 'S'
    most_one = df['Embarked'].value_counts().idxmax() 
    df['Embarked'] = df['Embarked'].fillna(most_one)

    #now let's replace the characters with the required digits
    my_dict = {"S": 0, "C": 1, "Q": 2} #we will use this dictionary in map function
    df['Embarked'] = df['Embarked'].map(my_dict)


    #3) Convert "Sex" column into numeric.
    #Replace "male" with 0 and replace "female" with 1. 
    #For the missing ones (if any), just fill with the most common one.

    #first, let's fill the missing ones with the most common one
    #the following code reveals the most common one is 'S'
    #As you may guess, the most common one is male passenger in Titanic.
    most_one = df['Sex'].value_counts().idxmax() 
    df['Sex'] = df['Sex'].fillna(most_one)
    my_dict = {"male": 0, "female": 1}
    df['Sex'] = df['Sex'].map(my_dict)


    #4) The column named "Cabin" includes strings with letters and numbers, 
    #for example "C85" and the letter refers to the cabin group. 
    #This column should indicate integer value of 1 for cabin name starts with letter "A". 
    #The rest of the numbers should appear according this dictionary: 
    #{"A": 1, "B": 2, "C": 3, "D": 4, "E": 5, "F": 6, "G": 7, "U": 8}. 
    #Any missing entries should be replaced with 0.

    my_dict = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5, 'F': 6, 'G': 7, 'U': 8}
    #the following code gives starting letters of each entry under Cabin column
    df['Cabin'] = df['Cabin'].str.extract('([A-Za-z0-9])')

    df['Cabin'] = df['Cabin'].map(my_dict)
    df['Cabin'] = df['Cabin'].fillna(0) #fill NaNs with 0s.
    df['Cabin'] = df['Cabin'].astype(int) #convert data type to integer

    #5) Fill missing values from "Age" column according to the following ad-hoc imputation technique: 
    #A random integer withdrawn from the set (mean - standard deviation, mean + standard deviation).


    #Remember we replaced NaNs with -1s in Question-1. 
    #Let's load back NaNs so that mean and standard deviation won't get affected
    ps_age = df['Age'].replace(-1, float('NaN'))
    #which ones are NaNs
    na_age = ps_age.isna()
    #mean and std of all ages
    mean_age = ps_age.mean()
    std_age = ps_age.std()
    #how many NaNs
    size_na_age = sum(na_age)

    #Let's use Numpy package to randomly withdraw integer values from the given range 
    #(mean - standard deviation, mean + standard deviation) as many as size
    import numpy as np
    #randint() gives an array which can be inserted in Pandas series
    ps_rand_age_array = np.random.randint(max(0, mean_age - std_age), mean_age + std_age, size = size_na_age)
    #Only replace for those NaNs.
    ps_age[na_age] = ps_rand_age_array
    #Make sure it is int type again.
    ps_age = ps_age.astype(int)
    df['Age'] = ps_age

    #6) Any missing values from "Fare" column should be replaced with 0.
    #Remember we replaced NaNs with -1s in Question-1. 
    #Let's load back NaNs so that mean and standard deviation won't get affected
    ps_fare = df['Fare'].replace(-1, float('NaN'))
    #which ones are NaNs
    na_fare = ps_fare.isna()
    #Only replace for those NaNs.
    ps_fare[na_fare] = 0
    #Make sure it is int type again.
    ps_fare = ps_fare.astype(int)
    df['Fare'] = ps_fare


    # 7) Update "Fare" column according to the following:
    # if 0 <= 'Fare' < 10, then 'Fare' = 0
    # if 10 <= 'Fare' < 20, then 'Fare' = 1
    # if 20 <= 'Fare' < 30, then 'Fare' = 2
    # if 30 <= 'Fare' < 100, then 'Fare' = 3
    # if 100 <= 'Fare' < 200, then 'Fare' = 4
    # if 200 <= 'Fare' then 'Fare' = 5
    df.loc[ df['Fare'] < 10, 'Fare'] = 0
    df.loc[(df['Fare'] >= 10) & (df['Fare'] < 20), 'Fare'] = 1
    df.loc[(df['Fare'] >= 20) & (df['Fare'] < 30), 'Fare'] = 2
    df.loc[(df['Fare'] >= 30) & (df['Fare'] < 100), 'Fare'] = 3
    df.loc[(df['Fare'] >= 100) & (df['Fare'] < 200), 'Fare'] = 4
    df.loc[(df['Fare'] >= 200), 'Fare'] = 5


    # 8) Update "Age" column according to the following:
    # if 'Age' <= 10, then 'Age' = 0
    # if 'Age' > 10 & 'Age' <= 15 then 'Age' = 1
    # if 'Age' > 15 & 'Age' <= 20 then 'Age' = 2
    # if 'Age' > 20 & 'Age' <= 25 then 'Age' = 3
    # if 'Age' > 25 & 'Age' <= 35 then 'Age' = 4
    # if 'Age' > 35 & 'Age' <= 40 then 'Age' = 5
    # if 'Age' > 40 & 'Age' <= 60 then 'Age' = 6
    # if 'Age' > 60 then 'Age' = 6
    df.loc[ df['Age'] <= 10, 'Age'] = 0
    df.loc[(df['Age'] > 10) & (df['Age'] <= 15), 'Age'] = 1
    df.loc[(df['Age'] > 15) & (df['Age'] <= 20), 'Age'] = 2
    df.loc[(df['Age'] > 20) & (df['Age'] <= 25), 'Age'] = 3
    df.loc[(df['Age'] > 25) & (df['Age'] <= 35), 'Age'] = 4
    df.loc[(df['Age'] > 35) & (df['Age'] <= 40), 'Age'] = 5
    df.loc[(df['Age'] > 40) & (df['Age'] <= 60), 'Age'] = 6
    df.loc[ df['Age'] > 60, 'Age'] = 6

    return df

#Call your function here
new_titanic = feature_extraction(new_titanic)
new_titanic.head()
df.head()
```

### Question 1
Which state has the most counties in it? (hint: consider the sumlevel key carefully! You'll need this for future questions too...)

*This function should return a single string value.*

```python
def my_function_question_1(args):
    df = args.copy()
    df = df.reset_index() #let's create a new column which is equal to row index labels in the dataframe
    #Notice that county names are labeled with the value of 50 under "SUMLEV" column.
    df = df[df['SUMLEV'] == 50] #select and create a new dataframe that has county names only
    df = df.set_index(['STNAME', 'CTYNAME']) #set both STNAME and CTYNAME as index of the new dataframe
    list_states = df.reset_index(level = 1).index #this will create repeated state names in an index type data structure
    list_counts = list_states.value_counts() #we can count the repeated state names
    most_county_state = list_counts.idxmax() #find the max one
    #we could accomplish the same task with a single line
    # most_county_state = df.reset_index(level=1).index.value_counts().idxmax()
    
    return most_county_state


my_function_question_1(census_df)
```

### Question 2
Only looking at the three most populous counties for each state, what are the three most populous states (in order of highest population to lowest population)? Use `CENSUS2010POP`.

*This function should return a list of string values.*

```python
def my_function_question_2(args):
    df = args.copy()
    df = df.reset_index() #let's make sure to keep the current index
    df = df[df['SUMLEV'] == 50] #again we are only interested in counties in each state
    df = df[['STNAME', 'CENSUS2010POP']] #create a new dataframe with STNAME and CENSUS2010POP
    df = df.sort_values(['STNAME','CENSUS2010POP'], ascending = False)
    df = df.groupby('STNAME').head(3) #tabulate states with 3 most populous counties 
    df = df.groupby("STNAME").sum() #sum populations 3 most populous counties within each state
    #sort states in order of highest population to lowest population
    df = df.sort_values('CENSUS2010POP', ascending = False)
    df = df.reset_index() #restore state index as a column
    my_states_list =  list(df['STNAME'].loc[0:2]) #select the first 3 states from the top
    
    return my_states_list


my_function_question_2(census_df)
```

### Question 3
Which county has had the largest absolute change in population within the period 2010-2015? (Hint: population values are stored in columns POPESTIMATE2010 through POPESTIMATE2015, you need to consider all six columns.)

e.g. If County Population in the 5 year period is 100, 120, 80, 105, 100, 130, then its largest change in the period would be |130-80| = 50.

*This function should return a single string value.*

```python
def my_function_question_3(args):
    df = args.copy()
    df = df.reset_index()
    df = df[df['SUMLEV'] == 50]
    list_pops = ['POPESTIMATE2010','POPESTIMATE2011','POPESTIMATE2012','POPESTIMATE2013',
                 'POPESTIMATE2014','POPESTIMATE2015']
    df['POPMIN'] = df.loc[:, list_pops].min(axis = 1)
    df['POPMAX'] = df.loc[:, list_pops].max(axis = 1)
    df['POPCHANGE'] = df['POPMAX'] - df['POPMIN']
    df = df[['CTYNAME', 'POPCHANGE']].sort_values('POPCHANGE', ascending = False).reset_index()
    county_name = df['CTYNAME'].loc[0]

    return county_name

my_function_question_3(census_df)
```

### Question 4
In this datafile, the United States is broken up into four regions using the "REGION" column. 

Create a query that finds the counties that belong to regions 1 or 2, whose name starts with 'Washington', and whose POPESTIMATE2015 was greater than their POPESTIMATE 2014.

*This function should return a 5x2 DataFrame with the columns = ['STNAME', 'CTYNAME'] and the same index ID as the census_df (sorted ascending by index).*

```python
def my_function_question_4(args):
    df = args.copy()
    df = df.reset_index()
    df = df[df['SUMLEV'] == 50]
    df = df[(df['REGION'] <= 2)]
    df = df[(df['POPESTIMATE2015'] > df['POPESTIMATE2014'])]
    #df['CTYNAME'] = df['CTYNAME'].str[0:10]
    df = df[(df['CTYNAME'].str[0:10] == 'Washington')].sort_index()
    cols_to_keep = ['STNAME', 'CTYNAME']
    df = df[cols_to_keep]
    
    return df

my_function_question_4(census_df)
```
