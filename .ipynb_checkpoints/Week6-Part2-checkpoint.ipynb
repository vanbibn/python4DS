#### Read from a SQL Database
Let's first take a look at the "flights.db" on online database processor: https://sqliteonline.com/<br/>
There are 3 tables in this database: "airlines", "airport" and "routes".<br/>
Database Link: https://sites.google.com/site/yasinunlu/home/research/new1/flights.db<br/>
This is a SQLite database file.

```python
import sqlite3
db_file = "data\\flights_db\\flights.db"

conn = None
try:
    conn = sqlite3.connect(db_file)
except Error as e:
    print(e)
    
type(conn)
```

### Cursor object

A cursor is a temporaty work-space/area for SQL commands (alocates a little bit of memory to work and then will return some data)

```python
# create cursor object
cur = conn.cursor()

# execute some SQL
cur.execute("SELECT * FROM airlines LIMIT 5;")

# # Iterate over the cursor
# for result in cur:
#     print(result)

# Fetch one result
# print(cur.fetchone())

# Fetch all results as list
rows = cur.fetchall()  # returns a list of tuples

print(type(rows))
print(type(rows[0]))


for row in rows:
    print(row)
```

Let's read the database table into a dataframe.

```python
# we already have a connection
df_airlines = pd.read_sql_query("SELECT * FROM airlines;", con = conn)

print(df_airlines.shape)

df_airlines.head()
```

```python
conn.close()
```

## Exercise:
When we look into airlines.json file, we notice that it is not in an indented format. We can convert this file into an indented file format. It will still be in .json file. But we will work on a smaller file "countries.json" to see how it actually works.<br/>
1) Read each .json file as a single string by using `read()` function in Python.<br/>
2) You must notice this is not in a nice format (i.e. no indentation). Create a JSON object by using `json.loads()`.<br/> 
3) Convert the JSON object into an indented format by using `json.dumps()` function.<br/>
4) Write the formatted data into a file with .json extension. Use `write()` function. For example, the JSON formatted data of df_tracks dataframe should be stored under a name "countries_.json".

```python
import json as JSON
```

```python
# 1. Read each .json file as a single string by using read() function in Python.
with open("data\\flights_db\\countries.json", "r") as json_file:
    json_data = json_file.read()

print(json_file.closed)
print(type(json_data))
json_data
```

```python
print(json_data)
```

```python
# 2. You must notice this is not in a nice format (i.e. no indentation). Create a JSON object by using json.loads().
json_object = JSON.loads(json_data)
print(json_object)
```

```python
# 3. Convert the JSON object into an indented format by using json.dumps() function.
json_formatted = JSON.dumps(json_object, indent = 2)

print(json_formatted)
```

```python
print(type(json_formatted))
```

```python
# 4. Write the formatted data into a file with .json extension. Use write() function. 
# For example, the JSON formatted data of df_tracks dataframe should be stored under a name "countries_.json".

with open("data\\flights_db\\countries_JSON.json", "w") as file:
    file.write(json_formatted)
```
