```python
#let's clean up the data
new_columns = ("summer_participations",
"summer_gold",
"summer_silver",
"summer_bronze",
"summer_total",
"winter_participations",
"winter_gold",
"winter_silver",
"winter_bronze",
"winter_total",
"total_participations",
"total_gold",
"total_silver",
"total_bronze",
"total_combined")
df.columns = new_columns
```

```python
df
```

```python
names_ids = df.index.str.split('\s\(') # split the index by '('

df.index = names_ids.str[0] # the [0] element is the country name (new index) 
```

```python
names_ids.str[1]
```

```python
#create a new col named "ID" with the 3 letter abbrev.
df['ID'] = names_ids.str[1].str[:3] # the [1] element is the abbreviation or ID (take first 3 characters from that)

df
```

```python
df = df.drop('Totals')  # drop last row "Totals"
df.head()
```

## Question-4
Which country has the biggest difference between their summer gold medal counts and winter gold medal counts relative to their total gold medal count? 

$$\frac{Summer~Gold - Winter~Gold}{Total~Gold}$$

Only include countries that have won at least 1 gold in both summer and winter.

You can use a build-in Pandas function.

```python
gold['diff_rel_tot'] = gold['diff'] / df['total_gold']
```

```python
gold['diff_rel_tot'].min()  # least difference btw summer and winter gold medal counts
```

```python
gold['diff_rel_tot'].idxmin()  # country with least difference
```

```python
gold.loc['Canada']
```

To find the country with the greatest relative difference (i.e. maximum) we need to only include countries that have at least 1 gold medal for both summer and winter. Otherwise, the country's `diff` = `total_gold` and the `diff_rel_tot` = 1

```python
gold[(gold['winter_gold'] > 0) & (gold['summer_gold'] > 0)]['diff_rel_tot'].idxmax()
```

```python
gold.loc['Bulgaria']
```

```python
gold[(gold['winter_gold'] > 0) & (gold['summer_gold'] > 0)].sort_values('diff_rel_tot', ascending=False)
```

What countries have more winter gold medals than summer gold medals? 

```python
gold[gold['winter_gold'] > gold['summer_gold']].sort_values('diff_rel_tot', ascending=False)
```
