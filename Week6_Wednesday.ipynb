## Exercises:
When we look into airlines.json file, we notice that it is not in an indented format. We can convert this file into an indented file format. It will still be in .json file. But we will work on a smaller file "countries.json" to see how it actually works.<br/>
1) Read each .json file as a single string by using `read()` function in Python.<br/>
2) You must notice this is not in a nice format (i.e. no indentation). Create a JSON object by using `json.loads()`.<br/> 
3) Convert the JSON object into an indented format by using `json.dumps()` function.<br/>
4) Write the formatted data into a file with .json extension. Use `write()` function. For example, the JSON formatted data of df_tracks dataframe should be stored under a name "countries_.json".

1) Read each .json file as a single string by using read() function in Python.

```python
my_file = open("countries.json", "r")
my_data = my_file.read()
my_file.close()
my_data #not a readable format!
```

2) You must notice this is not in a nice format (i.e. no indentation). Create a JSON object by using json.loads().

```python
import json as JSON
json_object = JSON.loads(my_data)
print(json_object) #still not pretty
```

3) Convert the JSON object into an indented format by using json.dumps() function.

```python
json_formatted_str = JSON.dumps(json_object, indent = 2) #we can print it in a nice format

print(json_formatted_str) 
```

4) Write the formatted data into a file with .json extension. Use write() function. For example, the JSON formatted data of df_tracks dataframe should be stored under a name "tracks_formatted.json".

```python
my_file = open("countries_.json","w")
my_file.write(json_formatted_str)
my_file.close()
```
